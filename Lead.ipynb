```python
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier,plot_tree


X = df.drop(["Converted"],axis=1)
Y = df["Converted"]
# feature extraction
model = DecisionTreeClassifier()
rfe = RFE(model,step=3)
fit = rfe.fit(X, Y)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
for feature in fit.support_.nonzero():
    print(feature)
    print(X.columns[feature])
```

```python
x=df[['Lead Number', 'Lead Origin', 'Lead Source', 'TotalVisits','Total Time Spent on Website', 'Last Activity', 'Country','Specialization', 'What is your current occupation', 'Tags','Lead Quality', 'Lead Profile', 'City', 'Asymmetrique Profile Index','Asymmetrique Activity Score', 'Asymmetrique Profile Score','Last Notable Activity']]
y=df["Converted"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score,recall_score,roc_curve, auc
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
x_pred=classifier.predict(X_train)
results=accuracy_score(y_train,x_pred)
print("Accuracy score of training data:",results*100)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy of testing data:",result2)
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
models=[]
accuracy=[]
models.append("Logistic Regression")
accuracy.append(result2)
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()

```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
x_pred=classifier.predict(X_train)
results=accuracy_score(y_train,x_pred)
print("Accuracy score of training data:",results*100)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy score of testing data:",result2*100)
models=[]
accuracy=[]
models.append("Decision Tree")
accuracy.append(result2)
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
train_scores=[]
test_scores=[]
values=[i for i in range(1,21)]
for i in values:
    model=DecisionTreeClassifier(max_depth=i)
    model.fit(X_train,y_train)
    train_yhat=model.predict(X_train)
    train_acc=accuracy_score(y_train,train_yhat)
    train_scores.append(train_acc)
    test_yhat=model.predict(X_test)
    test_acc=accuracy_score(y_test,test_yhat)
    test_scores.append(test_acc)
plt.plot(values,train_scores,'-o',label="Train")
plt.plot(values,test_scores,'-o',label="Test")
plt.legend()
plt.show()

    
```
